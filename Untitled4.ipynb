```python
import os
import sqlite3
from flask import Flask, request

app = Flask(__name__)

# 1. ثغرة SQL Injection
@app.route("/vulnerable_sql", methods=["GET", "POST"])
def vulnerable_sql():
    username = request.args.get("username")  # Input مباشر بدون فحص
    conn = sqlite3.connect("users.db")
    cursor = conn.cursor()
    # ثغرة: تنفيذ استعلام مباشرة باستخدام إدخال المستخدم
    query = f"SELECT * FROM users WHERE username = '{username}'"
    cursor.execute(query)
    result = cursor.fetchall()
    conn.close()
    return str(result)

# 2. ثغرة Command Injection
@app.route("/vulnerable_command", methods=["POST"])
def vulnerable_command():
    user_input = request.form.get("cmd")
    # ثغرة: تنفيذ أمر نظام باستخدام إدخال المستخدم
    os.system(f"echo {user_input}")
    return "Command executed!"

# 3. ثغرة Cross-Site Scripting (XSS)
@app.route("/vulnerable_xss", methods=["GET"])
def vulnerable_xss():
    name = request.args.get("name")  # إدخال المستخدم مباشرة في الإخراج
    return f"<h1>Hello, {name}</h1>"

# 4. Hardcoded Secrets
SECRET_KEY = "123456"  # ثغرة: مفتاح حساس مكتوب في الكود مباشرة
DB_PASSWORD = "password123"  # كلمة مرور حساسة مكتوبة في الكود

# 5. ثغرة Insecure Deserialization
@app.route("/vulnerable_deserialize", methods=["POST"])
def vulnerable_deserialize():
    import pickle
    serialized_data = request.data
    # ثغرة: فك تسلسل (deserialization) بدون فحص
    obj = pickle.loads(serialized_data)
    return f"Deserialized: {obj}"

# 6. ثغرة Path Traversal
@app.route("/vulnerable_path", methods=["GET"])
def vulnerable_path():
    filename = request.args.get("file")
    # ثغرة: قراءة ملفات باستخدام إدخال المستخدم
    with open(f"/var/www/{filename}", "r") as file:
        return file.read()

if __name__ == "__main__":
    app.run(debug=True)
```
